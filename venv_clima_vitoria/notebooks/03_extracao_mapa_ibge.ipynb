```python
import requests
import json
import os

# O flush=True obriga o VS Code a mostrar a mensagem na hora
print("🌍 Conectando à API do IBGE para buscar a malha territorial de Vitória...", flush=True)

url_ibge = "https://servicodados.ibge.gov.br/api/v3/malhas/municipios/3205309?formato=application/vnd.geo+json"

# Criando um "disfarce" para o Governo achar que somos um navegador normal
headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/120.0.0.0 Safari/537.36'
}

try:
    # Adicionamos o limite de 15 segundos (timeout) e o cabeçalho de disfarce (headers)
    response = requests.get(url_ibge, headers=headers, timeout=15)

    if response.status_code == 200:
        geojson_vitoria = response.json()
        
        caminho_pasta = '../data/raw'
        os.makedirs(caminho_pasta, exist_ok=True)
        caminho_salvamento = f'{caminho_pasta}/vitoria_limite_ibge.geojson'
        
        with open(caminho_salvamento, 'w', encoding='utf-8') as f:
            json.dump(geojson_vitoria, f, ensure_ascii=False)
            
        print(f"✅ Sucesso! O arquivo GeoJSON foi salvo em: {caminho_salvamento}")
        print("🎉 O esqueleto do nosso mapa interativo já está pronto para o Power BI!")
    else:
        print(f"❌ Falha ao buscar o mapa. O IBGE respondeu com o Status: {response.status_code}")

except requests.exceptions.Timeout:
    print("\n❌ O servidor do IBGE demorou mais de 15 segundos e cortamos a conexão. Podem estar em manutenção.")
except Exception as e:
    print(f"\n❌ Ocorreu um erro inesperado: {e}")
```
